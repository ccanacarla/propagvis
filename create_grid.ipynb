```python
import json

def make_grid_geojson(H=10, W=10, cell_size=3.0, out="grid_64x64.geojson"):
    feats = []
    for i in range(H):
        for j in range(W):
            x0, y0 = j * cell_size, i * cell_size
            x1, y1 = x0 + cell_size, y0 + cell_size
            cx, cy = (x0 + x1) / 2.0, (y0 + y1) / 2.0
            idx = i * W + j

            feats.append({
                "type": "Feature",
                "properties": {
                    "area": cell_size ** 2,       # Ã¡rea
                    "centroid_x": cx,
                    "centroid_y": cy,
                    "id": idx
                },
                "geometry": {
                    "type": "Polygon",
                    "coordinates": [[
                        [x0, y0],
                        [x1, y0],
                        [x1, y1],
                        [x0, y1],
                        [x0, y0]
                    ]]
                }
            })

    geojson = {
        "type": "FeatureCollection",
        "crs": {
            "type": "name",
            "properties": {
                "name": "urn:ogc:def:crs:OGC:1.3:CRS84"
            }
        },
        "features": feats
    }

    with open(out, "w", encoding="utf-8") as f:
        json.dump(geojson, f, ensure_ascii=False, indent=2)

# gerar exemplo
make_grid_geojson(H=20, W=20, cell_size=1.0, out="data/grid_64x64.geojson")
```
